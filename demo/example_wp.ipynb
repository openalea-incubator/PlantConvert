# Using openalea mtg and Weber&Penn, visualize and convert to different format

Example to create a `.mtg` file using `.mtg` and `weberpenn` openalea modules. We then visualize the plant and export to `.opf` format.

```python
import openalea.weberpenn as wp
import openalea.plantgl.all as pgl
import openalea.plantconvert as pc 
import openalea.mtg as mtg

plant_file = ("../data/A3B4.mtg")
```

```python
pc.Plant().mtg.display()
```

```python
plant = pc.plantconvert.Plant(file=plant_file)
plant.read()
```

```python
plant.mtg.display()
```

```python
%gui qt
scene = pc.geometry.get_scene(plant.mtg)
pgl.Viewer.display(scene)
```
